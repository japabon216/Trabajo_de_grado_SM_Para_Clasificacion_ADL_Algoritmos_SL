```python
dataframe_names = ['d1', 'd2', 'd3', 'd4', 'd5', 'd6', 'd7', 'd8', 'd9', 'd10', 'd11', 'd12', 'd13', 'd14', 'd15', 'd22', 'd23']

dataframes = [dataset1, dataset1_2, dataset2, dataset2_2, dataset3, dataset4, dataset5,
              dataset6, dataset8, dataset9, dataset11_corregido, dataset12teoria, dataset13,
              dataset13_2, dataset15,  dataset17, dataset20]

x_acc_values = [df['X_Acc'] for df in dataframes]

plt.figure(figsize=(10, 6))
sns.set(style="whitegrid")
sns.violinplot(data=x_acc_values, inner="quartiles")
plt.xticks(ticks=range(len(dataframe_names)), labels=dataframe_names, rotation=45)
plt.xlabel("DataFrames")
plt.ylabel("Valores de Aceleración (X_Acc)")
plt.title("Distribución de X_Acc en Diferentes DataFrames")
plt.tight_layout()
plt.savefig('/content/drive/MyDrive/TESIS_DATASET/Graficas de violin  /Despues_X.png', dpi=300, bbox_inches='tight')
plt.show()
```
