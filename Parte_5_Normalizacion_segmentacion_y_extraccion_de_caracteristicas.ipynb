##Entropia

```python
from scipy.stats import entropy

def calcular_entropia(datos):
    # Calcular las probabilidades de ocurrencia de cada valor
    unique_values, value_counts = np.unique(datos, return_counts=True)
    probabilities = value_counts / len(datos)

    # Calcular la entropía usando la fórmula de Shannon
    entropia = -np.sum(probabilities * np.log2(probabilities))

    return entropia

grupo_segmento = datasetfinal.groupby('segmento')

resultados = pd.DataFrame()

# Calcular la entropía para cada segmento y cada columna de interés
for nombre_segmento, grupo in grupo_segmento:
    entropia_resultado = {}
    entropia_resultado['Dataset'] = grupo['Dataset'].values[0]
    entropia_resultado['Activity'] = grupo['Activity'].values[0]
    entropia_resultado['Code segmento'] = nombre_segmento

    for columna in ['X_Acc', 'Y_Acc', 'Z_Acc', 'X_Gyro', 'Y_Gyro', 'Z_Gyro']:
        entropia_resultado[f'{columna}_entropy'] = calcular_entropia(grupo[columna])

    resultados = resultados.append(entropia_resultado, ignore_index=True)

resultados
```

##Promedio

```python
resultados = pd.DataFrame(columns=['Dataset', 'Activity', 'X_Acc_mean', 'Y_Acc_mean', 'Z_Acc_mean', 'X_Gyro_mean', 'Y_Gyro_mean', 'Z_Gyro_mean', 'Code', 'segmento'])

for segmento, segmento_df in datasetfinal.groupby('segmento'):

    promedios = segmento_df[['X_Acc', 'Y_Acc', 'Z_Acc', 'X_Gyro', 'Y_Gyro', 'Z_Gyro']].mean()

    # Tomar el valor de la primera fila para las columnas 'Dataset', 'Activity' y 'Code'
    first_row = segmento_df.iloc[0]

    # Crear una fila para el resultado
    resultado_fila = {
        'Dataset': first_row['Dataset'],
        'Activity': first_row['Activity'],
        'X_Acc_mean': promedios['X_Acc'],
        'Y_Acc_mean': promedios['Y_Acc'],
        'Z_Acc_mean': promedios['Z_Acc'],
        'X_Gyro_mean': promedios['X_Gyro'],
        'Y_Gyro_mean': promedios['Y_Gyro'],
        'Z_Gyro_mean': promedios['Z_Gyro'],
        'Code': first_row['Code'],
        'segmento': segmento
    }

    # Agregar la fila al DataFrame de resultados
    resultados = resultados.append(resultado_fila, ignore_index=True)

resultados
```

##Promedio absoluto

```python

# Función para calcular el promedio absoluto
def promedio_absoluto(segmento):
    return segmento.abs().mean()

# Agrupar por segmento y calcular el promedio absoluto
grouped = datasetfinal.groupby('segmento')
std_deviation = grouped[['X_Acc', 'Y_Acc', 'Z_Acc', 'X_Gyro', 'Y_Gyro', 'Z_Gyro']].apply(promedio_absoluto)

first_row = grouped[['Dataset', 'Activity', 'Code']].first()

result = first_row.join(std_deviation, on='segmento')

result.reset_index(inplace=True)

result.columns = ['segmento', 'Dataset', 'Activity', 'Code', 'X_Acc_abslt', 'Y_Acc_abslt', 'Z_Acc_abslt', 'X_Gyro_abslt', 'Y_Gyro_abslt', 'Z_Gyro_abslt']
```

##Promedio resultante

```python
# Función para calcular el promedio resultante
def calcular_promedio_resultante(segmento):
    sumatoria = np.sum(np.sqrt(segmento['X_Acc']**2 + segmento['Y_Acc']**2 + segmento['Z_Acc']**2))
    promedio = sumatoria / len(segmento)
    return promedio

resultados = datasetfinal.groupby('segmento').apply(calcular_promedio_resultante)

resultado_final = datasetfinal.groupby('segmento').first()[['Dataset', 'Activity', 'Code']]

resultado_final['promedio_resultante_Acc'] = resultados

# Para el giroscopio
def calcular_promedio_resultante_gyro(segmento):
    sumatoria = np.sum(np.sqrt(segmento['X_Gyro']**2 + segmento['Y_Gyro']**2 + segmento['Z_Gyro']**2))
    promedio = sumatoria / len(segmento)
    return promedio

resultados_gyro = datasetfinal.groupby('segmento').apply(calcular_promedio_resultante_gyro)
resultado_final['promedio_resultante_Gyro'] = resultados_gyro

resultado_final.reset_index(drop=True, inplace=True)
```

##Magnitud

```python
# Función para calcular la amgnitud
def calcular_magnitud(segmento):
    sumatoria = np.sum(np.sqrt(segmento['X_Acc']**2 + segmento['Y_Acc']**2 + segmento['Z_Acc']**2))
    return sumatoria

resultados = datasetfinal.groupby('segmento').apply(calcular_magnitud)

resultado_final = datasetfinal.groupby('segmento').first()[['Dataset', 'Activity', 'Code']]

resultado_final['energia_Acc'] = resultados

# Para el giroscopio
def calcular_magnitud_gyro(segmento):
    sumatoria = np.sum(np.sqrt(segmento['X_Gyro']**2 + segmento['Y_Gyro']**2 + segmento['Z_Gyro']**2))
    return sumatoria

resultados_gyro = datasetfinal.groupby('segmento').apply(calcular_magnitud_gyro)
resultado_final['energia_Gyro'] = resultados_gyro

resultado_final.reset_index(drop=True, inplace=True)
```
