```python
import random
import pandas as pd

def resample_dataframe(df):
    original_frequency = df['Frequency'].iloc[0]

    if original_frequency == 50:
        return df

    elif original_frequency > 50:
        new_rows = int(len(df) * (50 / original_frequency))
        drop_indices = random.sample(range(len(df)), len(df) - new_rows)

        new_df = df.drop(drop_indices).copy()

        new_df['Frequency'] = 50

        return new_df

    else:
        new_frequency = 50

        grouped = df.groupby('User')

        resampled_data = []

        for group_name, group_data in grouped:
            desired_rows = int(len(group_data) * (new_frequency / group_data['Frequency'].iloc[0]))

            repeated_group = pd.concat([group_data] * (desired_rows // len(group_data)))

            remaining_rows = desired_rows - len(repeated_group)
            if remaining_rows > 0:
                additional_rows = group_data.sample(n=remaining_rows, replace=True)
                repeated_group = pd.concat([repeated_group, additional_rows])

            ordered_group = repeated_group.sort_index()

            resampled_data.append(ordered_group)

        resampled_data = pd.concat(resampled_data, ignore_index=True)
        resampled_data['Frequency'] = 50
        return resampled_data
```

```python
# Funcion para ver los valores válidos de un acelerometro
import plotly.graph_objects as go
import pandas as pd

def grafico_acelerometro_valido(df):

    valores_nulos = df[['X_Acc', 'Y_Acc', 'Z_Acc']].isnull().any(axis=1).sum()
    valores_validos = len(df) - valores_nulos


    # Calcular los porcentajes
    porcentaje_nulos = (valores_nulos / len(df)) * 100
    porcentaje_validos = (valores_validos / len(df)) * 100

    labels = ['Valores nulos', 'Valores válidos']
    values = [porcentaje_nulos, porcentaje_validos]

    fig = go.Figure(data=[go.Pie(labels=labels, values=values)])

    fig.update_layout(
        title="Porcentaje de valores nulos y valores válidos en el acelerómetro"
    )

    fig.show()
```

```python
# Valores máximos y mínimos en el acelerometro
import plotly.graph_objects as go
import pandas as pd

def grafico_acelerometro_max_min(df):

    columnas = ['X_Acc', 'Y_Acc', 'Z_Acc']
    df_columnas = df[columnas]

    media = df_columnas.mean()
    mediana = df_columnas.median()
    maximo = df_columnas.max()
    minimo = df_columnas.min()

    desplazamiento = (media.max() - media.min()) * 0.05

    fig = go.Figure()

    fig.add_trace(go.Bar(
        x=columnas,
        y=media,
        name='Media',
        marker_color='rgba(55, 128, 191, 0.7)'
    ))

    fig.add_trace(go.Bar(
        x=columnas,
        y=mediana,
        name='Mediana',
        marker_color='rgba(219, 64, 82, 0.7)'
    ))

    for columna, valor_max, valor_min in zip(columnas, maximo, minimo):
        fig.add_annotation(
            x=columna,
            y=media.max() + desplazamiento,
            text="Máximo: " + str(valor_max),
            showarrow=False,
            font=dict(color='green')
        )
        fig.add_annotation(
            x=columna,
            y=media.min() - desplazamiento,
            text="Mínimo: " + str(valor_min),
            showarrow=False,
            font=dict(color='purple')
        )

    fig.update_layout(
        title="Media, Mediana y Etiquetas de Máximo y Mínimo de las columnas del acelerometro",
        xaxis_title="Columnas",
        yaxis_title="Valor"
    )

    fig.show()
```

## Preparación del Dataset número 1.

Debido a que este en este dataset todos sus archivos CSV tienen informoación necesaria, pero que afectan el precesamiento en sus primeras 40 filas, se borran las 40 filas de los 746 archivos de la carpeta de UMAFALL_Dataset.

```python
# Ruta de la carpeta con los archivos CSV
folder_path = "/content/drive/MyDrive/TESIS_DATASET/1_UMAFall_Dataset/UMAFall_Dataset"

# Ruta de la carpeta donde se guardarán los archivos corregidos
new_folder_path = "/content/drive/MyDrive/TESIS_DATASET/1_UMAFall_Dataset/UMAFall_Dataset_Corregido"

# Nombre del header que se buscará
header_name = "% TimeStamp"

# Porcentaje de similitud mínimo para considerar que se encontró el header
header_similarity = 0.6

# Contador de archivos exitosos
successful_files = 0

# Contador de archivos con errores
error_files = 0

# Número máximo de archivos con errores permitidos
max_error_files = 5

# Recorremos todos los archivos de la carpeta
for filename in sorted(os.listdir(folder_path)):

    if filename.endswith(".csv"):

        file_path = os.path.join(folder_path, filename)

        new_filename = "new_" + filename

        new_file_path = os.path.join(new_folder_path, new_filename)

        try:
            with open(file_path, "r") as f:
                reader = csv.reader(f, delimiter=";")
                rows = list(reader)

                header_found = False
                for i, row in enumerate(rows):
                    similarity = sum([word in row[0] for word in header_name.split()]) / len(header_name.split())
                    if similarity >= header_similarity:
                        header_found = True
                        break

                if not header_found:
                    raise ValueError("No se encontró el header adecuado")

                # Borramos las filas anteriores al header
                rows = rows[i:]

            with open(new_file_path, "w", newline="") as f:
                writer = csv.writer(f, delimiter=";")
                writer.writerows(rows)

            successful_files += 1

        except Exception as e:
            # Si ocurre un error, incrementamos el contador de archivos con errores
            error_files += 1

            # Si se alcanza el límite máximo de errores, detenemos el programa
            if error_files >= max_error_files:
                print("Se alcanzó el límite máximo de archivos con errores permitidos")
                break

            # Si no, mostramos el mensaje de error y continuamos con el siguiente archivo
            print(f"Error en el archivo {filename}: {str(e)}")
            continue

print(f"{successful_files} archivos fueron exitosamente creados del número total de archivos que es {len(os.listdir(folder_path))}")

```

El dataset presento un problema y es que algunos valores de acelerometro y giroscopio se leian equivocadamente, tomaba una coma como signo de mil, lo cual no era correcto, se propuso el siguiente código como solución:
Adicionalmente, como un primer vistazo se cambia las unidades del acelerometro de g a ms2

```python
def convert_decimal(val):
    # Si el valor es un flotante, lo devolvemos tal cual
    if isinstance(val, float):
        return val

    # Reemplaza la coma por un punto si el valor está en notación científica
    if 'E' in val or 'e' in val:
        val = val.replace(',', '.')

    # Divide el string en dos partes usando el primer punto como separador
    parts = val.split('.', 1)

    # Si solo hay una parte, no había ningún punto, por lo que devolvemos el número tal cual
    if len(parts) == 1:
        return float(parts[0])

    # Si hay dos partes, eliminamos los puntos adicionales de la segunda parte y las unimos de nuevo
    return float(parts[0] + '.' + parts[1].replace('.', ''))

# Ahora aplicamos esta función a las columnas
df_200['X_Acc'] = df_200['X_Acc'].apply(convert_decimal)
df_200['Y_Acc'] = df_200['Y_Acc'].apply(convert_decimal)
df_200['Z_Acc'] = df_200['Z_Acc'].apply(convert_decimal)

# Ahora realiza la conversión de las unidades.
g = 9.81
df_200['X_Acc'] = df_200['X_Acc'] * g
df_200['Y_Acc'] = df_200['Y_Acc'] * g
df_200['Z_Acc'] = df_200['Z_Acc'] * g
```

Debido a que en el dataset 2 cada usuario podia usar su frecuencia propia de celular sin realizar ninguna modificación se presentan inconvenientes y es que para trabajar con un dataset sin conocer la frecuencia es casi que imposible, por esto se propone leer las filas que hay en cada segundo para cada acticvidad de cada usuario.

El proceso que se realiza a continuación consiste en calcular la frecuencia de cada usuario, una vez calculadas las frecuencias estas serán asignadas al acc. Este proceso se realizará para el archivo del giroscopio.

Se seleccionan unicamente las columnas necesarias para que el proceso se realice más rápido

```python
columnas_deseadas = ['username','timestamp','activity']
acchc = acch[columnas_deseadas]
```

```python
# Crear una nueva columna 'second' que contenga únicamente el valor de los segundos de cada registro
acch['second'] = acch['timestamp'].dt.second
```

```python
# Agrupar los datos por usuario y actividad
grouped_data = acchc.groupby(['username', 'activity'])

# Calcular la frecuencia para cada grupo
result = []
for group_name, group_data in grouped_data:
    username, activity = group_name
    freq = group_data.resample('S', on='timestamp').size()
    freq_nonzero = freq[freq != 0]
    freq_mean = np.mean(freq_nonzero)
    freq_median = np.median(freq_nonzero) if len(freq_nonzero) > 0 else None
    freq_mode = freq_nonzero.mode().values[0] if len(freq_nonzero) > 0 else None
    result.append((username, activity, freq, freq_mean, freq_median, freq_mode))

for username, activity, freq, freq_mean, freq_median, freq_mode in result:
    print(f"La frecuencia para la actividad {activity} en el username {username} es:")
    print(freq.head(40))
    print("...")
    print(freq.tail(40))
    print(f"La frecuencia media para la actividad {activity} en el username {username} es: {freq_mean}")
    print(f"La frecuencia mediana para la actividad {activity} en el username {username} es: {freq_median}")
    print(f"La frecuencia moda para la actividad {activity} en el username {username} es: {freq_mode}")
    print()
```

El código anterior brinda información tanto de la media como de la mediana y moda, en casos donde estos 3 valores no sean similares se decidio no incluir estos valores, por eso se realiza lo siguiente, donde solo se agregan las frecuencias "Estables" y posteriormente se eliminan las filas con frecuencia vacia

```python
import os
import pandas as pd
import scipy.io

ruta_origen = '/content/drive/MyDrive/TESIS_DATASET/3_The_USC_SIPI_Human_Activiry_Dataset/USC-HAD'
ruta_destino = '/content/drive/MyDrive/TESIS_DATASET/3_The_USC_SIPI_Human_Activiry_Dataset/USC-HAD/Subject1_14_CSV'

# Obtener la lista de carpetas de Subject1 a Subject14 en la ruta de origen
carpetas = [x for x in os.listdir(ruta_origen) if x.startswith('Subject')]

# Imprimir el número de archivos por carpeta en la ruta de origen
for carpeta in carpetas:
    ruta_carpeta = os.path.join(ruta_origen, carpeta)
    archivos_mat = [x for x in os.listdir(ruta_carpeta) if x.endswith('.mat')]
    print(f'La carpeta {carpeta} contiene {len(archivos_mat)} archivos .mat')

# Procesar cada carpeta y transformar archivos .mat a archivos .csv
for carpeta in carpetas:
    ruta_carpeta = os.path.join(ruta_origen, carpeta)
    archivos_mat = [x for x in os.listdir(ruta_carpeta) if x.endswith('.mat')]
    ruta_destino_carpeta = os.path.join(ruta_destino, f'{carpeta}_CSV')

    # Crear la carpeta de destino si no existe
    if not os.path.exists(ruta_destino_carpeta):
        os.makedirs(ruta_destino_carpeta)

    # Procesar cada archivo .mat en la carpeta y transformarlo a .csv
    num_archivos = 0
    for archivo in archivos_mat:
        ruta_origen_archivo = os.path.join(ruta_carpeta, archivo)
        data = scipy.io.loadmat(ruta_origen_archivo)
        sensor_readings = data['sensor_readings']
        subject = data['subject'][0]
        activity = data['activity'][0]
        trial = data['trial'][0]
        df = pd.DataFrame(sensor_readings, columns=['x_ace', 'y_ace', 'z_ace', 'x_giro', 'y_giro', 'z_giro'])
        df.insert(0, 'subject', subject)
        df.insert(1, 'activity', activity)
        df.insert(2, 'trial', trial)
        nombre_archivo = f'new_{archivo[:-4]}.csv'
        ruta_destino_archivo = os.path.join(ruta_destino_carpeta, nombre_archivo)
        df.to_csv(ruta_destino_archivo, index=False)
        num_archivos += 1

    print(f'Se crearon {num_archivos} archivos de {len(archivos_mat)} en la carpeta {carpeta}_CSV')

```

Como se puede ver se crearon 60 nuevos arvhicos CSV a partir de los archivos .mat de cada una de las carpetas de los sujetos anteriores, a continuación se obtendrá un Dataset único por el USC_HAD_Dataset

```python
# Ruta donde se encuentran las carpetas Subject1, Subject2, ..., Subject14
folder_path = '/content/drive/MyDrive/TESIS_DATASET/3_The_USC_SIPI_Human_Activiry_Dataset/USC-HAD/Subject1_14_CSV/'

df_list = []

num_files = 0
num_rows = 0

# Iteramos sobre todas las carpetas de la ruta folder_path
for folder_name in os.listdir(folder_path):

    current_folder_path = os.path.join(folder_path, folder_name)

    if not os.path.isdir(current_folder_path):
        continue

    # Obtenemos los nombres de los archivos CSV en la carpeta
    files = os.listdir(current_folder_path)
    files = [f for f in files if f.endswith('.csv')]

    # Ignoramos la carpeta si no contiene archivos CSV
    if len(files) == 0:
        continue

    # Incrementamos el número total de archivos procesados
    num_files += len(files)

    temp_df_list = []

    # Iteramos sobre los archivos CSV de la carpeta actual
    for file in files:
        file_path = os.path.join(current_folder_path, file)
        temp_df = pd.read_csv(file_path, delimiter=';')
        temp_df_list.append(temp_df)

    # Concatenamos los DataFrames de la carpeta y los agregamos a la lista principal
    folder_df = pd.concat(temp_df_list, ignore_index=True)
    df_list.append(folder_df)
    print(f'{len(files)} archivos de la carpeta {folder_name} procesados exitosamente.')

# Concatenamos todos los DataFrames en uno solo
USC_HAD_Dataset = pd.concat(df_list, ignore_index=True)

# Guardamos el dataset completo como un archivo CSV
USC_HAD_Dataset.to_csv('/content/drive/MyDrive/TESIS_DATASET/3_The_USC_SIPI_Human_Activiry_Dataset/USC-HAD/USC_HAD_Dataset.csv', index=False)

print(f'Total de {num_files} archivos procesados y {len(USC_HAD_Dataset)} filas en el nuevo dataset.')
```

```python
import re

source_dir = '/content/drive/MyDrive/TESIS_DATASET/3_The_USC_SIPI_Human_Activiry_Dataset/USC-HAD/Subject1_14_CSV/'

dest_file = '/content/drive/MyDrive/TESIS_DATASET/3_The_USC_SIPI_Human_Activiry_Dataset/USC-HAD/USC_HAR_Definitivo_intento2.csv'

subject_dirs = [d for d in os.listdir(source_dir) if os.path.isdir(os.path.join(source_dir, d))]

# Ordenar los directorios de Subject en función del número
subject_dirs.sort(key=lambda d: int(re.search(r'\d+', d).group()))

dataframes = []

for dir in subject_dirs:

    dir_path = os.path.join(source_dir, dir)

    csv_files = [f for f in os.listdir(dir_path) if f.endswith('.csv')]

    # Ordenar los archivos CSV en función del número después de 'a' y luego después de 't'
    csv_files.sort(key=lambda f: (int(re.search(r'a(\d+)', f).group(1)), int(re.search(r't(\d+)', f).group(1))))

    # Recorrer cada archivo CSV y agregarlo a la lista de dataframes
    for file in csv_files:

        file_path = os.path.join(dir_path, file)

        # Leer el archivo CSV y agregarlo a la lista de dataframes
        df = pd.read_csv(file_path)
        dataframes.append(df)

# Concatenar todos los dataframes
result = pd.concat(dataframes)

# Escribir el dataframe resultante en el archivo de destino
result.to_csv(dest_file, index=False)
```

El dataset se presenta mediante 19 archivos txt los cuales representan a cada sujeto, dento de estos archivos se encuentran los datos raw de los 4 sensores así como su etiquetado en orden, esto es importante puesto que al momento de unir todos los archivos hay que extraer el valor del sujeto y ordenar por sujeto,arctividad y código, para que el orden se mantenga. Sin embargo esto de debe preguntar porque para el timestamp probablemente sea mejor dejar las filas como estan para adicionar este factor con la ayuda de la frecuencia.

```python

import os
import pandas as pd

folder_path = "/content/drive/MyDrive/TESIS_DATASET/6. Daliac/Archivos_Raw"

# Obtener la lista de archivos ordenados
file_list = sorted([file for file in os.listdir(folder_path) if file.endswith('.txt')])

# Crear una lista para almacenar los DataFrames
dfs = []

# Iterar sobre los archivos y agregar la columna 'User'
for i, file_name in enumerate(file_list):
    file_path = os.path.join(folder_path, file_name)
    df = pd.read_csv(file_path, delimiter=',')
    header = ['X_Acc_W', 'Y_Acc_W', 'Z_Acc_W', 'X_Gyro_W', 'Y_Gyro_W', 'Z_Gyro_W',
    'X_Acc_C', 'Y_Acc_C', 'Z_Acc_C', 'X_Gyro_C', 'Y_Gyro_C', 'Z_Gyro_C',
    'X_Acc_H', 'Y_Acc_H', 'Z_Acc_H', 'X_Gyro_H', 'Y_Gyro_H', 'Z_Gyro_H',
    'X_Acc_A', 'Y_Acc_A', 'Z_Acc_A', 'X_Gyro_A', 'Y_Gyro_A', 'Z_Gyro_A',
    'Aticity']
    df.columns = header
    df['User'] = i + 1  # Asignar el valor del nombre como la columna 'User'
    dfs.append(df)

# Concatenamos los archivos resultantes
concatenated_df = pd.concat(dfs, axis=0)
```
