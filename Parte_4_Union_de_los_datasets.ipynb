#Función

```python
def menores(df, x, y):
  count_x_acc = (df['X_Acc'] < -x).sum()

  count_y_acc = (df['Y_Acc'] < -x).sum()

  count_z_acc = (df['Z_Acc'] < -x).sum()

  count_x_gyro = (df['X_Gyro'] < -y).sum()

  count_y_gyro = (df['Y_Gyro'] < -y).sum()

  count_z_gyro = (df['Z_Gyro'] < -y).sum()

  print(f"Valores menores de {x} en X_Acc: {count_x_acc}")
  print(f"Valores menores de {x} en Y_Acc: {count_y_acc}")
  print(f"Valores menores de {x} en Z_Acc: {count_z_acc}")
  print(f"Valores menores de {y} en X_Gyro: {count_x_gyro}")
  print(f"Valores menores de {y} en Y_Gyro: {count_y_gyro}")
  print(f"Valores menores de {y} en Z_Gyro: {count_z_gyro}")
```

```python
def mayores(df, x,y):
  count_x_acc = (df['X_Acc'] > x).sum()

  count_y_acc = (df['Y_Acc'] > x).sum()

  count_z_acc = (df['Z_Acc'] > x).sum()

  count_x_gyro = (df['X_Gyro'] > y).sum()

  count_y_gyro = (df['Y_Gyro'] > y).sum()

  count_z_gyro = (df['Z_Gyro'] > y).sum()

  print(f"Valores mayores de {x} en X_Acc: {count_x_acc}")
  print(f"Valores mayores de {x} en Y_Acc: {count_y_acc}")
  print(f"Valores mayores de {x} en Z_Acc: {count_z_acc}")
  print(f"Valores mayores de {y} en X_Gyro: {count_x_gyro}")
  print(f"Valores mayores de {y} en Y_Gyro: {count_y_gyro}")
  print(f"Valores mayores de {y} en Z_Gyro: {count_z_gyro}")
```

```python
def eliminacion(df,x,y):
  if (y==0):
    df = df[
        (df['X_Acc'] < x) & (df['X_Acc'] > -x) &
        (df['Y_Acc'] < x) & (df['Y_Acc'] > -x) &
        (df['Z_Acc'] < x) & (df['Z_Acc'] > -x)
        ]
    return df
  elif (x==0):
    df = df[
        (df['X_Gyro'] < y) & (df['X_Gyro'] > -y) &
        (df['Y_Gyro'] < y) & (df['Y_Gyro'] > -y) &
        (df['Z_Gyro'] < y) & (df['Z_Gyro'] > -y)
        ]
    return df
  else:
    df = df[
        (df['X_Acc'] < x) & (df['X_Acc'] > -x) &
        (df['Y_Acc'] < x) & (df['Y_Acc'] > -x) &
        (df['Z_Acc'] < x) & (df['Z_Acc'] > -x) &
        (df['X_Gyro'] < y) & (df['X_Gyro'] > -y) &
        (df['Y_Gyro'] < y) & (df['Y_Gyro'] > -y) &
        (df['Z_Gyro'] < y) & (df['Z_Gyro'] > -y)
        ]
    return df
```
